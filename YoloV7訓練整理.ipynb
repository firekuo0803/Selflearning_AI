###Ë§áË£ΩËàá‰øÆÊîπÊâÄÈúÄÊ™îÊ°à

```python
#Ë§áË£Ω
import os
os.system("pip install ruamel.yaml")

data3 = "data"
cfg3 = "cfg"

data_path = f"/content/gdrive/MyDrive/{filenames}/yolov7/data/{data3}.yaml"
cfg_path = f"/content/gdrive/MyDrive/{filenames}/yolov7/cfg/training/{cfg3}.yaml"

d = open(data_path,"a")
c = open(cfg_path,"a")

d.write("")
d.close()
c.write("")
c.close()

import shutil
src_data=rf"/content/gdrive/MyDrive/{filenames}/yolov7/data/coco.yaml"
des_data=rf"/content/gdrive/MyDrive/{filenames}/yolov7/data/{data3}.yaml"
shutil.copy2(src_data, des_data)

src_cfg=rf"/content/gdrive/MyDrive/{filenames}/yolov7/cfg/training/{model}.yaml"
des_cfg=rf"/content/gdrive/MyDrive/{filenames}/yolov7/cfg/training/{cfg3}.yaml"
shutil.copy2(src_cfg, des_cfg)

#‰øÆÊîπdata
from ruamel.yaml import YAML

labels = [ 'jack sparrow' ]

yaml = YAML()
#data
with open(data_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
data["train"] = "./data/train"
with open(data_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)

#valid
with open(data_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
data["val"] = "./data/valid"
with open(data_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)

#nc
with open(data_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
data["nc"] = len(labels)
with open(data_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)

#labels
with open(data_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
data["names"] = labels
with open(data_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)

with open(data_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
  del data["download"]
with open(data_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)

with open(data_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
  del data["test"]
with open(data_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)

#‰øÆÊîπcfg
with open(cfg_path, "r", encoding='utf-8') as f:
  data = yaml.load(f)
data["nc"] = len(labels)
with open(cfg_path, "w", encoding="utf-8") as f:
    yaml.dump(data, f)
```
